```python
from helpers.read_data_tools import read_data, imageDataset, save_imageDataset, latentDataset, save_latentDataset, build_latentDataset
from helpers.pre_trained_autoencoder import load_autoencoder, encode, decode
```

```python
train_latentDataset = build_latentDataset(ImageDataset = train_imageDataset, autoencoder = autoencoder, batch_size = 32)
save_latentDataset(train_latentDataset, save_name = "train_latentDataset", compressed = False)
```
